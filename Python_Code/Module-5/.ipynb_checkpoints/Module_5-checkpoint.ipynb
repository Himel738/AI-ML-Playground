## Tuples in Python - Collection of Data (immutable)

```python
# Declaration
tup = (10,20,30,40)
float_tup = (10.5,10.2)

print(type(tup), tup)
print(type(float_tup),float_tup)

mixed_tup = (10.2,10, "str", True)
print(mixed_tup)

tup1 = (10) # It can't be a Tuple
print(type(tup1))

tup1 = (10,) # By a Comma We can Declare as a Tupple else it will be an integer
print(type(tup1))

tup1 = tuple([10,20]) # Tuple can be use as a list .. list can be used as tuple
print(type(tup1))

```

## Access

```python
tup = (10,20,30,10.5,"str",False)
print(tup[4])

# Slicing
new_tup = tup[0:3]
print(new_tup)
```

## immutability - can't be change declared variable

```python
tup = (10,20,30)
tup[1] = 40 #Can't
```

## Set In Python - Unique Element, Mutable, Unordered(indexing, and Slicing Not Avilable), Duplicate Not Allowed

```python
A = {1,2,3}
print(type(A), A)
```

```python
B = {}
print(type(B))
B = set()
print(type(B))
```

```python
# Access
S = {1, 2, 3, 4, 1, 2}
if 3 in S:
    print("3 Is Present")
else:
    print("Not Present")
for element in S:
    print(element)
```

## Sets Method - https://www.w3schools.com/python/python_ref_set.asp

```python
S = {5, 1, 2, 4, 3}
print(S)

S.add(10)
S.add(0)
print(S)

S.pop() # Deletinf value from Start
S.pop()
print(S)

S.remove(5)
print(S)
```

## Union, Intersection, Disjoint - Set Mathmatical Method

```python
a = {1,2,3}
b = {1,2,3,4}

print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b)) # Any common point
print(a.issubset(b))
```

## Dictionary - (Key (As a Key We Can Take Int, Float, String, Tuple) Value), Mutable, Unordered(indexing, and Slicing Not Avilable)

```python
dic = {}
print(type(dic))
```

```python
dic = {"name" : "KutuMia", "age" : "25", "address" : "Dhaka" , "numbers" : [10,20,30], "age" : 30} # If there is duplicate Key the latest will be print
print(type(dic), dic)
```

```python
# Access
print(dic["name"])
print(dic.get("age"))

dic["name"] = "KutuMia_"
print(dic)

print(dic.get("marks")) # But if you want a value instead of None
print(dic.get("marks", 0)) # Do This
```

## Dictionary Method - https://www.w3schools.com/python/python_ref_dictionary.asp

```python
dic = {"name" : "KutuMia", "age" : "25", "address" : "Dhaka" , "numbers" : [10,20,30], "age" : 30}
print(dic)

dic["math_marks"] = 30
print(dic)

dic.update({"english_marks" : 40})
print(dic)
```

```python
# Deleting

del dic["math_marks"]
print(dic)
```

```python
# Copy
dic_2 = dic.copy()
print(dic_2)
```

```python
# Testing

dic_3 = {[10,20,30] : "a"} # It will not work because Key should be immutable
dic_3 = {(10,20,30) : "a"} # It will work because key is immutable
```

## View Method of Dictionary

```python
dic = {"name" : "KutuMia", "age" : "25", "address" : "Dhaka" , "numbers" : [10,20,30], "age" : 30}
keys = dic.keys()
print(keys)

dic["math_numbers"] = 35
print(keys) # We didn't update Keys but the numbers updated automatically so thats why its real time update

values = dic.values()
print(values)

items = dic.items()
print(items)
```

```python
# Iteration

for key,value in dic.items():
    print(key,value)
```

## Dictionary Comprehension - {expression(Key : Value) , iterable , Condition}

```python
square = {x : x**2 for x in range(1,11) if x % 2 == 0} # x * 2 = x * 2 but x ** 2 = x to the power 2
print(square)
```

```python
co_ordinates = [(10,10.5), (20.5, 192), (101,102)]
locations = ["dhaka" , "chattogram" , "sylhet"]
exact_location = {co_or : loc for co_or, loc in zip(co_ordinates,locations)} # By Zip We can Combine Two list as a Dictionary
print(exact_location)
```

## 1. Given a List of Numbers, Make A list With unique Value

```python
numbers = [42, 7, 19, 3, 25, 11, 42, 11]
unique_values = set(numbers)
numbers = list(unique_values)
print(numbers)
```

## 2. Given a String , Print The Frequency of the Words

```python
text = "ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn ai data model train ai learn neural data ai model train learn"
words = text.split()

count ={}

for word in words:
    count[word] = count.get(word,0) + 1
print(count)
```
