## Tuples in Python - Collection of Data (immutable)

```python
# Declaration
tup = (10,20,30,40)
float_tup = (10.5,10.2)

print(type(tup), tup)
print(type(float_tup),float_tup)

mixed_tup = (10.2,10, "str", True)
print(mixed_tup)

tup1 = (10) # It can't be a Tuple
print(type(tup1))

tup1 = (10,) # By a Comma We can Declare as a Tupple else it will be an integer
print(type(tup1))

tup1 = tuple([10,20]) # Tuple can be use as a list .. list can be used as tuple
print(type(tup1))

```

## Access

```python
tup = (10,20,30,10.5,"str",False)
print(tup[4])

# Slicing
new_tup = tup[0:3]
print(new_tup)
```

## immutability - can't be change declared variable

```python
tup = (10,20,30)
tup[1] = 40 #Can't
```

## Set In Python - Unique Element, Mutable, Unordered(indexing, and Slicing Not Avilable), Duplicate Not Allowed

```python
A = {1,2,3}
print(type(A), A)
```

```python
B = {}
print(type(B))
B = set()
print(type(B))
```

```python
# Access
S = {1, 2, 3, 4, 1, 2}
if 3 in S:
    print("3 Is Present")
else:
    print("Not Present")
for element in S:
    print(element)
```

## Sets Method - https://www.w3schools.com/python/python_ref_set.asp

```python
S = {5, 1, 2, 4, 3}
print(S)

S.add(10)
S.add(0)
print(S)

S.pop() # Deletinf value from Start
S.pop()
print(S)

S.remove(5)
print(S)
```

## Union, Intersection, Disjoint - Set Mathmatical Method

```python
a = {1,2,3}
b = {1,2,3,4}

print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b)) # Any common point
print(a.issubset(b))
```
