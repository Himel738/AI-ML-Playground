## Min and Max

```python
inp = input()
numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])
Z = int(numbers[2])

min = x
max = x
if y < min:
    min = y
if Z < min:
    min = Z
if y > max:
    max = y
if Z > max:
    max = Z

print(min,max)

```

### Even, Odd, Positive and Negative

```python
n = int(input())
inp = input()
numbers = inp.split()

positive = 0
negative = 0
even = 0
odd = 0

for i in range(n):
    x = int(numbers[i])
    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1

    if x % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)

```
