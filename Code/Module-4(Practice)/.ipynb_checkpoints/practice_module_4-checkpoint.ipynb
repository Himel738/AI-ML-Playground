## Problem-01 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b1-smart-light-pilot-threshold

```python
inp  = input()
numbers = inp.split()
brightness = float (numbers[0])
threshold = float(numbers[1])

if brightness >= threshold:
    print("ON")
else:
    print("OFF")
    
```

## problem-02 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b3-sensor-calibrator-min-max-scaling

```python
inp  = input()
numbers = inp.split()
x = float (numbers[0])
min_v = float (numbers[1])
max_v = float (numbers[2])

norm = float (x - min_v) / (max_v - min_v)

print(f"{norm :.2f}")



```

## problem-03 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b4-mini-loss-tracker-average-monitor

```python
n = int (input())
sum = 0.0;
targate = 0;

for i in range(n+1):
    x = float (input())
    list_numbers = []
    list_numbers.append(x)

    if(i == 0):
        targate = x
    else:
        sum += x
sum = sum / n
if sum <= targate:
    print(f"PASS")
else:
    print(f"RETRY")
    
    
```

## problem - 04 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b5-majority-vote-gate-simple-ensemble

```python
n = int (input())
count_yes = 0
count_no = 0
for i in range(n):
    x = str (input())
    if x == "YES":
        count_yes += 1
    else:
        count_no += 1
if count_yes >= count_no:
    print("ACCEPT")
else:
    print("REJECT")
```

## PROBLEM - 05 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/ai-mood-detector

```python
string = str (input())

if "happy" in string or "joy" in string or "smile" in string:
    print(f"Happy Mood")
elif "sad" in string or "cry" in string or "angry" in string:
    print(f"Sad Mood")
else:
    print(f"Neutral Mood")
    
```

## Problem - 06 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a2-ai-brightness-analyzer

```python
inp = input()
numbers = inp.split()
sum = 0

for num in numbers:
    sum = sum + int (num)
avg = sum / float (len(numbers))

if avg < 85:
    print("Dark Image")
elif 85 <= avg <= 170:
    print("Normal Image")
else:
    print("Bright Image")
```

## Problem - 07 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a3-ai-bias-detector

```python
string = input()
char = string.split()
count_a = 0
count_b = 0
for c in char:
    if c == 'A':
        count_a +=1
    else:
        count_b +=1
accuracy_a = (count_a / float(count_a + count_b)) * 100
accuracy_b = (count_b / float(count_a + count_b)) * 100

if accuracy_a > 70 or accuracy_b > 70:
    print("Biased Model")
else:
    print("Fair Model")
```

## Problem - 08 - https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a4-ai-keyword-counter
