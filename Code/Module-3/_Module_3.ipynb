## Indexing and Slicing

```python
string = "hello world"
print(string[0])
print(string[7])
```

## for Slicing - String_Name [start : end : step ]  by default step is 1

```python
first_message = message [0:24]
print(first_message)
```

```python
first_message = message [0:24:2]
print(first_message)
```

## In Backward Indexing - The last World Will be indexed as -1

```python
print(first_message[-1])
print(first_message[-2])
print(first_message[-3])
print(first_message[-4])
print(first_message[-5])
```

## String Methods - https://www.w3schools.com/python/python_ref_string.asp

```python
string = "Welcome To Ml Playground"

processed_string = string.lower() #we can also use Upper

#length
print(len(string))

#Find a Word
print("playground" in processed_string)

#FIND THE substring Index from the begining if two playground prsent..first one will be the ans
shuru_index = processed_string.find("playground")
print(shuru_index)

 #From the last
shesh_index = processed_string.rfind("playground")
print(shesh_index)

#count the substring
count = processed_string.count("playground")
print(count)

#Replace the Substring
new_string = string.replace("Ml", "AI/Ml") #after changing it has to be stored in a new_string...because the orginal string is immutable
print(new_string)

```
