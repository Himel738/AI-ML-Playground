## Filtering On String

```python
import pandas as pd
data = {
    "Name": [
        "Rahim Uddin", "Karim Hasan", "Ayesha Akter", "Nusrat Jahan",
        "Tanvir Ahmed", "Sadia Islam", "Imran Hossain", "Farhan Rahman",
        "Mitu Sultana", "Shakib Al Hasan", "Rafiq Islam", "Jannat Ara",
        "Arif Khan", "Sumaiya Rahman", "Mahmudul Hasan"
    ],
    "City": [
        "Dhaka", "Chattogram", "Khulna", "Rajshahi",
        "Sylhet", "Barishal", "Rangpur", "Dhaka",
        "Chattogram", "Khulna", "Rajshahi", "Sylhet",
        "Barishal", "Rangpur", "Dhaka"
    ],
    "Department": [
        "IT", "HR", "Finance", "Marketing",
        "IT", "Finance", "HR", "Marketing",
        "IT", "Finance", "HR", "Marketing",
        "IT", "Finance", "HR"
    ],
    "Salary": [
        55000, 48000, 62000, 53000,
        70000, 58000, 45000, 60000,
        67000, 75000, 47000, 52000,
        69000, 61000, 50000
    ]
}
df = pd.DataFrame(data)
df
```

```python
df.loc[df['City'].str.contains("Dh")]
```

```python
df.loc[df['City'].str.contains("dh", case=False)] # Case Sensitivity
```

```python
df.loc[df['City'].str.contains(r"ch", case=False)] # Regular Expression- ^ch means only starts with ch
```

```python
# end with 'ka'
df.loc[df['City'].str.contains(r"ka$", case=False)]
```

```python
df.loc[df['Name'].str.contains(r"^[AEIOU]", case=False)]
```

```python
df.loc[df['City'].str.contains(r"Dh|Ch", case=False)]
```
