# 🔹 2D Line Plot

A **2D line plot** is a graph used to show the relationship between two variables by connecting data points with a continuous line.

---

## 🔹 Components
- **X-axis (Independent Variable)** → Input or sequence (e.g., time)  
- **Y-axis (Dependent Variable)** → Output or measured values  
- **Data Points** → Individual observations  
- **Line** → Connects points to show trends  

---

## 🔹 Key Features
- Shows trends and patterns clearly  
- Easy to understand and interpret  
- Useful for continuous data  

---

## 🔹 Common Uses
- Time series analysis  
- Sales or growth tracking  
- Temperature or performance monitoring  

---

## 🔹 Why It’s Important
- Helps visualize changes over time  
- Makes comparisons easier  
- Useful in Exploratory Data Analysis (EDA)  

---

## 🔹 Summary

A 2D line plot helps visualize the **relationship between two variables**, making trends and patterns easy to understand.
