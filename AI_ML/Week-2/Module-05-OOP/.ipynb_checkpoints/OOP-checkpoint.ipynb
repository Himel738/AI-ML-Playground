# 🔹 Class Attribute in Python

A **class attribute** is a variable that is shared among all instances (objects) of a class.  
It is defined inside the class but **outside any method**.

---

## 🔹 Key Points
- Shared by all objects of the class  
- Defined at class level  
- Saves memory (no duplication per object)  
- Can be accessed using both **class name** and **object**

---

## 🔹 Instance vs Class Attribute

| Feature            | Class Attribute        | Instance Attribute      |
|--------------------|----------------------|-------------------------|
| Defined In         | Class                | Constructor (`__init__`)|
| Shared             | Yes                  | No                      |
| Access             | Class & Object       | Object only             |
| Memory Usage       | Efficient            | More                    |

---

## 🔹 Summary

- Class attributes are **common for all objects**
- Instance attributes are **unique for each object**

````python
class phone:
    category = "Electronics"    
# creating object
apple = phone()
blueberry = phone()
motorola = phone()

#print(type(apple))
print(apple.category)
print(blueberry.category)
print(motorola.category)
````

````python
class phone:
    category = "Electronics"
    
# creating object
apple = phone()
apple.category = "Super Electronics" # Changing the Default Value
blueberry = phone()
motorola = phone()

#print(type(apple))
print(apple.category)
print(blueberry.category)
print(motorola.category)
````

# 🔹 Constructor in Python (OOP)

A **constructor** is a special method that is automatically called when an object is created from a class.  
It is mainly used to **initialize object attributes**.

---

## 🔹 Key Points
- Automatically executed during object creation  
- Used to assign initial values to object properties  
- Helps ensure objects start in a valid state  
- Defined using a special method name  

---

## 🔹 Types of Constructors
- **Default Constructor** → No parameters (except self)  
- **Parameterized Constructor** → Accepts values during object creation  

---

## 🔹 Why Constructor is Important
- Reduces repetitive code  
- Makes object creation easier  
- Improves readability and structure  

---

````python
class phone:
    category = "Electronics"

    #constructor
    def __init__(self, model, battery, camera, battery_percentage = 100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage
    # Method
    def charge(self,hour):
        print(f"Charge Completed by {hour}")
    def capture(self, photo):
        if(self.battery_percentage) <= 0:
            print("No Charge")
        else:
            self.battery_percentage -= photo
            print(f"Photo Captured in {self.model}")
class cooling_Mechanism:
    def __init__(self,cooling_method):
        self.cooling_method = cooling_method

    def cooling_on(self):
        print(f"The System is Being Cooled by {self.cooling_method}")
    
# creating object
apple = phone("i-phone - 17" , 3000, 40)
blueberry = phone("B-17", 4000, 30)
motorola = phone("M-17", 3500, 35)

print(apple.camera)
print(motorola.camera)
print(motorola.battery_percentage)

apple.capture(10)
motorola.capture(15)

print(apple.battery_percentage)
print(motorola.battery_percentage)

````

# 🔹 Inheritance in OOP (Python)

**Inheritance** is a concept where one class (child class) acquires the properties and behaviors of another class (parent class).

---

## 🔹 Key Points
- Promotes **code reusability**
- Helps create a **hierarchical relationship**
- Child class can use and extend parent class features
- Reduces code duplication

---

## 🔹 Types of Inheritance
- **Single Inheritance** → One parent, one child  
- **Multiple Inheritance** → One child, multiple parents  
- **Multilevel Inheritance** → Chain of inheritance  
- **Hierarchical Inheritance** → Multiple children from one parent  

---

## 🔹 Advantages
- Reuse existing code  
- Easier maintenance  
- Improves readability  
- Supports method overriding (runtime polymorphism)

---

## 🔹 Important Terms
- **Parent Class (Base Class)** → The class being inherited  
- **Child Class (Derived Class)** → The class that inherits  
- **Method Overriding** → Child class modifies parent method  

---

## 🔹 Summary

Inheritance allows a class to **reuse and extend another class**, making programs more organized and efficient.

## Single Inheritance

````python
class SmartPhone(phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor
    def charge(self, hour):
        print("Fast Charging in Process")
        super().charge(hour)
        
pro = SmartPhone("X", 5000, 100, "SnapDragon")
print(pro.model)
pro.charge(1)
````

## Multiple Inheritance

````python
class SmartPhone_CollingMode(SmartPhone, cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        SmartPhone.__init__(self, model, battery, camera, processor)
        cooling_Mechanism.__init__(self, cooling_method)
pro_cooling = SmartPhone_CollingMode("Y", 5000, 100, "SD", "Nitrogen")
print(pro_cooling.processor) # SmartPhone Class
print(pro_cooling.battery) # Phone Class
print(pro_cooling.cooling_method) # Cooling Mechanism Class
print(pro_cooling.cooling_on()) # Cooling Mechanism Class Method
print(pro_cooling.charge(1)) # SmartPhone er modified charge phone class theke inherit
````

# 🔹 Polymorphism in OOP

**Polymorphism** means “many forms”.  
It allows the same method or function name to behave differently depending on the object.

---

## 🔹 Key Points
- Same method name, different behavior  
- Improves flexibility and code reusability  
- Works with different classes using a common interface  

---

## 🔹 Types of Polymorphism
- **Compile-time Polymorphism** → Method Overloading  
- **Runtime Polymorphism** → Method Overriding  

---

## 🔹 Why It’s Important
- Reduces code complexity  
- Makes programs more scalable  
- Helps write cleaner and more maintainable code  

---

## 🔹 Summary

Polymorphism allows one interface to be used for **multiple implementations**, making code more flexible and powerful.

````python
class Camera:
    def __init__(self, name):
        self.name = name
    def capture(self):
        print(f"A Photo is Captured")
````

````python
class Smart(Camera):
    def __init__(self,name,resoulution):
        super().__init__(name)
        self.resoulution = resoulution
    def capture(self):
        print(f"A Photo is Captured by {self.name}")
        
class DSLR(Camera):
    def __init__(self,name,resoulution):
        super().__init__(name)
        self.resoulution = resoulution
    def capture(self):
        print(f"A Photo is Captured by {self.name}")
        
class Drone(Camera):
    def __init__(self,name,resoulution):
        super().__init__(name)
        self.resoulution = resoulution
    def capture(self):
        print(f"A Photo is Captured by {self.name}")
        
phone = Smart("Phone", 30)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 150)

print(phone.capture())
print(dslr.capture())
print(drone.capture())
````

# 🔹 Encapsulation in OOP

**Encapsulation** is the process of wrapping data (variables) and methods (functions) into a single unit (class) and restricting direct access to some of the data.

---

## 🔹 Key Points
- Combines data and methods in one class  
- Protects data from outside interference  
- Controls access using access modifiers  
- Achieves data hiding  

---

## 🔹 Access Modifiers
- **Public** → Accessible from anywhere  
- **Protected** → Accessible within class and subclasses  
- **Private** → Accessible only inside the class  

---

## 🔹 Why It’s Important
- Improves data security  
- Prevents accidental modification  
- Makes code more organized and maintainable  

---

## 🔹 Summary

Encapsulation ensures **data hiding and controlled access**, making programs safer and more structured.

````python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei ## Data is Now Private
    def charge(self):
        print("Phone is Charging")
    def imei_getter(self):
        return self.__imei
Iphone = Mobile("Phone", "17", "1xkaf1")

print(Iphone.imei_getter())
        
````

# 🔹 Abstraction in OOP

**Abstraction** is the process of hiding implementation details and showing only the essential features of an object.

---

## 🔹 Key Points
- Focuses on *what an object does*, not how it does it  
- Hides complex logic from the user  
- Achieved using abstract classes and methods  
- Improves code simplicity and readability  

---

## 🔹 Why It’s Important
- Reduces complexity  
- Enhances code clarity  
- Makes systems easier to maintain and extend  

---

## 🔹 Real-Life Example
Using a mobile phone:
You can make calls or take photos without knowing how the internal system works.

---

## 🔹 Summary

Abstraction hides unnecessary details and exposes only the **essential functionality**, making programs easier to use and understand.

````python
from abc import ABC, abstractmethod
class Telephone(ABC):# ABC -> Abstract Base Class
    @abstractmethod
    def make_call(self):
        pass
    
class Sphone(Telephone):
    def make_call(self):
        print("Makin a Call With SPhone")
class Iphone(Telephone):
    def make_call(self):
        print("Makin a Call With IPhone")

ip = Iphone()
print(ip.make_call())
````
