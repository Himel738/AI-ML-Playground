# 🔹 Streamlit

**Streamlit** is an open-source Python library used to build interactive web applications for data science and machine learning projects.

---

## 🔹 Key Points
- Turns Python scripts into web apps easily  
- No need for HTML, CSS, or JavaScript  
- Great for data visualization and dashboards  
- Fast and beginner-friendly  

---

## 🔹 Features
- Interactive widgets (buttons, sliders, inputs)  
- Real-time updates  
- Easy integration with data science libraries  
- Simple and clean UI  

---

## 🔹 Why It’s Popular
- Very easy to use  
- Saves development time  
- Ideal for showcasing ML models and projects  

---

## 🔹 Summary

Streamlit allows you to **build and share interactive web apps using only Python**, especially for data science and ML.

# 🔹 Input Elements in Streamlit

**Input elements** in Streamlit are used to take user input interactively in a web app.

---

## 🔹 Common Input Elements

- **Text Input** → Take text from user  
- **Number Input** → Take numeric values  
- **Text Area** → Multi-line text input  
- **Select Box** → Choose one option from list  
- **Multi Select** → Choose multiple options  
- **Radio Button** → Select one option  
- **Checkbox** → True/False input  
- **Slider** → Select value from a range  
- **Button** → Trigger actions  


````python
import streamlit as st
st.title("Input your information", anchor = False)
st.divider()

name = st.text_input("Enter Your Name:",placeholder="Type your name here")
# st.write("Your name is:", name)

st.divider()

age = st.number_input("Enter Your Age:", min_value=0, max_value=120, step=1, value = None, placeholder="Type your age here")
pressed = st.button("Enter to Confirm", type="primary")

if pressed:
    st.write(f"Your name is: {name} and Your age is: {age}")

st.divider()

# password = st.text_input("Enter Your Password:", type="password", placeholder="Type your password here")
# st.write("Your password is:", password)
````

# 🔹 Media Elements in Streamlit

**Media elements** in Streamlit are used to display multimedia content like images, audio, and video in a web app.

---

## 🔹 Types of Media Elements

- **Image** → Display images  
- **Audio** → Play audio files  
- **Video** → Play video content  

---

## Images

````python
import streamlit as st

st.title("Input Your Files", anchor = False)
st.divider()

images = st.file_uploader("Enter Your Image",
                 type=["jpg", "jpeg", "png"],
                  accept_multiple_files=True
                )

if images:
  col = st.columns(len(images))
  for i,image in enumerate(images):
    with col[i]: # By With We Open the Column
      st.image(image)
````

## Audio

````python
import streamlit as st
st.title("Input Your Files-", anchor = False)
st.divider()

st.audio("audio/tomal.mpeg", loop = True)

audios = st.file_uploader("Enter Your Audio Files",
                 type=["mp3", "wav", "mpeg"]
                )

if audios:
  st.audio(audios)
````

## Video

````python
import streamlit as st
st.title("Input Your Files-", anchor = False)
st.divider()

st.video("video/bike.mp4", loop = True)

videos = st.file_uploader("Enter Your Video Files",
                 type=["mp4", "avi", "mov"]
                )
button = st.button("Submit")

if button:
  if videos:
    st.video(videos)
    st.success("Video uploaded successfully!")
  else:
    st.error("Please upload a video file before submitting.")
````

# 🔹 API (Application Programming Interface)

An **API** is a set of rules that allows different software applications to communicate with each other.

---

## 🔹 Simple Idea
An API works like a **messenger**:
- You send a request  
- The API processes it  
- It returns a response  

---

## 🔹 Key Points
- Connects different systems or applications  
- Allows data exchange between software  
- Hides internal implementation details  
- Makes development faster and easier  

---

## 🔹 Real-Life Example
- You use a weather app  
- The app requests data from a weather server using an API  
- The API returns temperature, humidity, etc.  

---

## 🔹 Types of APIs
- **Web API** → Works over the internet (most common)  
- **REST API** → Uses HTTP methods (GET, POST, etc.)  
- **Library API** → Functions provided by a library  

---

## 🔹 Why APIs Are Important
- Saves development time  
- Enables integration between services  
- Allows reuse of functionality  

---

## Implementation of GEMINI API - .env file is Secrates

````python
from google import genai
import os
from dotenv import load_dotenv
import streamlit as st

load_dotenv()
api_key = os.environ.get("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)

response = client.models.generate_content(
  model = "gemini-3-flash-preview",
  contents = "Tell me a joke"
)

print(response.text)

st.markdown(response.text)
````
