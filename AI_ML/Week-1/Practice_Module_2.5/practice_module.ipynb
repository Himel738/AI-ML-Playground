## problem - 01: Write a function introduce(name, age) that prints:

 My name is <name> and I am <age> years old.
 
Call it using positional arguments.

Sample Input:
Rhim 28
Sample Output:
My name is Rahim and I am 28 years old.


```python
def introduce(name,age):
    print(f"My name is {name} and I am {28} years old.")
inp = input()
words = inp.split()
name = words[0]
age = int (words[1])
introduce(name,age)
```

## problem - 02: Write a Python generator function called batch_generator. It should take a list of numbers (representing data samples) and a batch_size. The generator should yield one batch at a time as a list.

Example Input:
data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3

Expected Output (when iterated):
[1, 2, 3]
[4, 5, 6]
[7, 8]


```python
def batch_generator(numbers, batch_size):
    for i in range(0, len(numbers), batch_size):
        yield numbers[i : i + batch_size]
inp = input("data = ")
numbers = list(map(int,inp.split())) # map() applies a function to each item in a sequence.
batch_size = int (input("batch_size = "))
x = batch_generator(numbers, batch_size)
for batch in x:
    print(batch)
```

## Problem - 03 - Given a list [1,2,3,4], use map() and lambda to create a new list with squares of each number.

Sample Input:
 No input

Sample Output:
[1, 4, 9, 16]


```python
def square_of_list(lst):
    #
    lst = map(lambda x : x**2, lst)
    return list(lst) # Because map() does NOT return a list directly return an object or can be said iterator
lst = [1, 2, 3, 4]
modified_lst = square_of_list(lst)
print(modified_lst)
```

## problem - 04 - Write a function power(base, exponent) that returns the result of base raised to exponent. If the exponent is not given, it should calculate the square.
Sample Input:
5
Sample Output:
25
Sample Input 2:
5 3
Sample Output 2:
125


```python
def power(base, expotent):
    #
    if(expotent == None):
        return base*base
    return base**expotent
inp = input()
numbers = inp.split()
base = int (numbers[0])
if len(numbers) > 1:
    expotent = int (numbers[1])
else:
    expotent = None

print(power(base,expotent))
```

 ## problem - 05 - Given a list [1,2,3,4,5,6], use filter() to select even numbers and map() with lambda to square them.
Sample Input:
 No input
Sample Output:
[4, 16, 36]


```python
def square_even(lst):
    lst = map(lambda y : y*y ,filter(lambda x : x % 2 == 0, lst))
    return list(lst)
    
lst = [1, 2, 3, 4, 5, 6]
modified_list = square_even(lst)
print(modified_list)
```
