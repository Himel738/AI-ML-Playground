#  File in Machine Learning Perspective

File is used to store data permanently (in secondary memory).

##  Common file formats in Machine Learning

- **CSV (.csv)** → tabular data (most common)  
- **JSON (.json)** → structured data (APIs, configs)  
- **Excel (.xlsx)**  
- **Images (.jpg, .png)** → for computer vision  
- **Text files (.txt)** → NLP tasks  

---

##  Summary

From ML perspective, a file is:

- **Input** → dataset  
- **Processing unit** → cleaned/transformed data  
- **Output** → trained model & results  
- **Reusable resource** → saved for future predictions  

## File Read and Write Expression

Syntax:
open("filename", mode)

Types of Modes:

Read Mode (r)
- Read entire file
- File must exist

Write Mode (w)
- Old content is deleted
- New content is saved
- Creates file if not exists

Append Mode (a)
- Old content remains
- New content is added at the end
- Creates file if not exists

````python
file = open("sample.txt", "r")
content = file.read()
print(content)
print(type(content))
print(file.closed)

file.close() # We Have to close file or there is chance to leekage Data
print(file.closed)
````

````python
file = open("sample.txt", "r")
content = file.readlines() # It Will Create List depends on New Line
print(content)
print(type(content))
````

````python
with open("sample.txt", "r") as file: # By a with Keyword we can close file automatically
    content = file.readlines()
    content = list(map(str.strip, content))
    filter_content = filter(lambda x : len(x) >= 1, content)
    print(list(filter_content))
print(file.closed)
````

````python
with open("sample.txt", "r") as file: 
    for line in file: # Extra New Line For Data And One for Print Function ..Double NewLine
        print(line, end ="") # No newline is added ...Output stays on the same line
print(file.closed)
````

````python
with open("sample.txt", "r") as file: 
    for line in file: # Extra New Line For Data And One for Print Function ..Double NewLine
        l = line.strip() # removes leading and trailing whitespace (spaces, tabs, newline) from the string line and stores the cleaned result in l.
        print(l)
print(file.closed)
````

## File Write and Append

````python
with open("test.txt", "w") as file: # Create a New File
    file.write("Hello Test File\n")
    file.write("How Learning is Going")
````

````python
with open("test.txt", "w") as file: # Writing On a Existing File it will delete all the previous information
    file.write("Second Attempt\n")
    file.write("Learning is going Fast\n")
````

````python
# Append - it will not Overwrite Data
with open("test.txt", "a") as file: # Writing On a Existing File it will Not delete all the previous information
    file.write("Learning is going Fast of Ai/Ml\n")
    file.write("Its All About Fun\n")
````

````python
strings = ["Hello", "Hi", "Good Bye", "Whats App"]
with open("test.txt", "a") as file: # Writing On a Existing File it will Not delete all the previous information
    file.writelines(strings)
    
````

## File Pointer - In Python, a file pointer is the current position of the cursor inside a file from where data is read or written.

````python
with open("sample.txt", "r") as file:
    print(file.tell()) # Staring Index of the File
    print(file.read())
    print(file.tell()) # Last index of the file

    print(file.read()) # due to read from last index ...it dosen't print anything
````

````python
with open("sample.txt", "r") as file:
    print(file.read(5))
    print(file.tell())
    print(file.read())
    print(file.tell())
````

````python
# Seek() Function
with open("sample.txt", "r") as file:
    print(file.tell())
    print(file.read(5))
    print(file.tell())

    file.seek(0)
    print(file.tell())
    print(file.read(5))
````

## Practice Problem 1 :

Get the text from sample.txt , and count the following


Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

````python
from functools import reduce
with open("sample.txt", "r") as file:
    strings = file.readlines()
    total_lines = len(strings)
    print(total_lines)

    # Print Total Number of Words
    number_of_words = map(lambda x : len(x.split()), strings) # map() → applies a function to each element separately and returns a list (or iterator)
    total_number_of_words = reduce(lambda x,y : x + y, number_of_words) # reduce() → combines all elements into one single value
    print(total_number_of_words)

    # Number of Characters-Space and New Line are not characters So we have to delete it
    string_list = list(map(str.strip, strings)) # new line delete
    string_list = list(map(lambda x : x.replace(" ", ""), string_list)) #Remove Spaces
    number_of_char = list(map(lambda x : len(x) , string_list)) # Char in every line
    total_number_of_char = reduce(lambda x,y : x+y, number_of_char)
    print(total_number_of_char)

    with open("counter_of_string.txt", "w") as file:
        file.write(f"Total Line : {total_lines}\nTotal Numbers of Words : {total_number_of_words}\nTotal Number of Characters: {total_number_of_char}\n")
````

## Practice Problem 2:
Write then read with one open

### File Modes Table

| Mode | Read | Write | File Must Exist | Old Data |
|------|------|-------|-----------------|----------|
| r+   | Yes  | Yes   | Yes             | Stays    |
| w+   | Yes  | Yes   | No              | Deleted  |
| a+   | Yes  | Yes   | No              | Stays    |

````python
with open("write_read.txt", "w+") as file:
    file.write("Hello World")
    print(file.tell())
    file.seek(0)
    print(file.read()) # As cursor in last index..so no print
    file.truncate(5) # Removes The String
    file.seek(0)
    print(file.read())
````

## Exception Handling in Python

Exception handling is used to handle runtime errors.

---

### Basic Syntax

```python
try:
    # code that may cause error
except:
    # runs if error occurs
```

---

### Example

```python
try:
    x = int(input("Enter number: "))
    print(10 / x)
except:
    print("Error occurred")
```

---

### Multiple Exceptions

```python
try:
    x = int(input())
    print(10 / x)
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

---

### else Block

```python
try:
    x = int(input())
except:
    print("Error")
else:
    print("No error occurred")
```

---

### finally Block

```python
try:
    f = open("file.txt")
except:
    print("Error")
finally:
    print("This will always run")
```

---
