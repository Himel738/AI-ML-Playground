## Basic Structure of Function - Declaration - def name(parameters)
                                               "-----> Doc String ----->"
                                               Instruction
                                               return type                         

```python
def welcome_message(): # No Parameter, No Return Type Function
    print("Hello User")
welcome_message()
```

```python
def welcome_message(user): 
    print(f"Hello {user}")
welcome_message("KutuMia")
```

## If User Dosen't Give Input

```python
def welcome_message(user = "None"): # Set a Default Value
    print(f"Hello {user}")
welcome_message()
```

## Summation Of Two Number Through Fuction and Return the result

```python
def adder(a,b):
    """
    Add Two Numbers
    a ---> int
    b ---> int
    """
    sum = a + b
    return sum
m = adder(10,30) * 5
print(m)
```

## Positional Arguments - When You are Passing value to Funcion called Argument, and The value goes in function called parameters
                       Types- 1. Positional
                              2. Keyword

```python
def adder(a,b,c): # a is mapping with - 10, b mapping With - 20, c mapping with - 30 thats called positional arguments 
    print(f"{a},{b},{c}")
adder(10,20,30)
```

## Variable Sized Positional Arguments

```python
def adder(*args): 
    print(f"{args}")
adder(10,20,30,40)
```

```python
def adder(*args): # We can Pass So Many Variables
    for i in args:
        print(f"{i}")  
adder(10,20,30,40)
```

## Keyword Arguments 

```python
def adder(a,b,c): 
    print(f"{a},{b},{c}")
adder(a=10, c=20, b=30)
```

## Variable Sized Keyword Arguments

```python
def adder(**args):
    print(args)
adder(a=10, b=30, c=40, n_estimator=50)
```

```python
def adder(**args):
    for key,value in args.items():
        print(f"Key : {key} , Value : {value}")     
adder(a=10, b=30, c=40, n_estimator=50)
```

## Return Type of a Function

```python
def welcome_message(): # No Parameter, No Return Type Function
    print("Hello User")
print(type(welcome_message()))
```

```python
def adder(a,b):
    """
    Add Two Numbers
    a ---> int
    b ---> int
    """
    sum = a + b
    return sum
print(type(adder(10,20)))
```

```python
def adder(a,b):
    """
    Add Two Numbers
    a ---> int
    b ---> int
    """
    sum = a + b
    return sum
print(type(adder(10,20.0)))
```

```python
def adder(a,b):
    """
    Add Two Numbers
    a ---> int
    b ---> int
    """
    sum = a + b
    return sum
print(type(adder(10.20,20.20)))
```

```python
def welcome_message(user): 
    return f"Hello {user}"
print(type(welcome_message("KutuMia")))
```

## Return Multiple Value

```python
def adder(a,b):
    return a,b
print(type(adder(10.20,20.20)))

# Unzip
x,y = adder(10.20,20.20)
print(f"{x} , {y}")
```

## Iterator - A List can be accesed by index but a set or dictionary? can't so we can use iterator

```python
s = {10,20,30,40,50} # A Set
s_iter = iter(s)
print(type(s_iter))

print(next(s_iter)) # As Set is not ordered... and can't Access by index
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))

print(f"Again")

for i in s: # internally created iterator...and running untill ending
    print(f"{i}")

```

## Generator - you have a memory 4 GB but Given data set is 10 GB... its not possible to load the data set..the system will crash... we can use chunk... where generator helps or it can be called yield that can pass value by chunk

```python
lst = [x for x in range(500)] # Dataset
#print(lst)

# Chunk
def data_loader(chunk_size,lst):
    for i in range(0, len(lst), chunk_size):
        yield lst[i: i + chunk_size]
x = data_loader(5,lst)
print(type(x))
print(next(x))
print(next(x))

    
```

## Lamda - inline and anonymonus.. def: lembda variable a , variable b: Expression, Map and Apply Function

```python
import pandas as pd

df = pd.DataFrame({'temperature(F)': [32,212,98.6,68]})
df
df['temperature(C)'] = df['temperature(F)'].apply(lambda x : (x - 32) * 5 / 9)
print(df)
```

```python
import pandas as pd
df = pd.DataFrame({'A' : [1, 2, 3, 4] , 'B' : [5, 6, 7, 8]})
df
df['A'] = df['A'].apply(lambda x : x ** 0.5)
df
```

```python
import pandas as pd
df = pd.DataFrame({'total_spent' : [120, 600, 450, 800]})
df
df['is_high_value'] = df['total_spent'].apply(lambda x : 1 if x > 500 else 0)
df
```
