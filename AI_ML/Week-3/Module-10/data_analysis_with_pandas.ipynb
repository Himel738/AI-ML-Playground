It helps work with structured data like tables (rows & columns).

---

## 🔹 Key Features
- Easy data handling and cleaning  
- Supports tabular data (like Excel sheets)  
- Fast and efficient operations  
- Built on top of NumPy  

---

## 🔹 Core Data Structures
- **Series** → 1D labeled array  
- **DataFrame** → 2D table (rows & columns)  

---

## 🔹 Why Use Pandas?
- Analyze large datasets easily  
- Perform filtering, grouping, and aggregation  
- Handle missing data  
- Widely used in Data Science & ML  

---

## 🔹 Summary

Pandas is used to **load, process, and analyze data efficiently in Python**.

# 🔹 Working With Different File Types in Pandas

Pandas allows you to **read and write data in multiple file formats**, making it very powerful for data analysis.

---

## 🔹 Common File Types

### 📄 CSV (Comma Separated Values)
- Most commonly used format  
- Easy to read and write  

```python
df = pd.read_csv("data.csv")
df.to_csv("output.csv", index=False)
```

---

### 📊 Excel Files
- Used in business and reporting  

```python
df = pd.read_excel("data.xlsx")
df.to_excel("output.xlsx", index=False)
```

---

### 🗂️ JSON Files
- Used for web APIs and nested data  

```python
df = pd.read_json("data.json")
df.to_json("output.json")
```

---

### 🗃️ Text Files
- Custom delimiter-based files  

```python
df = pd.read_csv("data.txt", delimiter="\t")
```
---
## Parquet File

```python
parquet_file = pd.read_parquet('File_name.parquet')
print(parquet_file)
```
---

### 🗄️ SQL Databases
- Read data directly from databases  

```python
import sqlite3

conn = sqlite3.connect("data.db")
df = pd.read_sql("SELECT * FROM table_name", conn)
```

---

## 🔹 Key Points
- Supports multiple file formats  
- Easy import and export  
- Handles large datasets efficiently  
- Works well with real-world data sources  

---

## 🔹 Why It’s Important
- Data comes in different formats  
- Pandas helps unify and process them easily  
- Essential for data analysis workflows  

---

## 🔹 Summary

Pandas allows you to **work with various file types like CSV, Excel, JSON, and databases**, making data handling flexible and efficient.

## Data Frame From inbuilt Data with Columns and index

````python
my_list =[['Alice', 25], ['Bob', 30], ['Charlie', 40]]
list_df = pd.DataFrame(my_list, columns = ['Names', 'Age'], index = [1, 2, 3])
list_df
````

````python
my_tuple =(('Alice', 25), ('Bob', 30), ('Charlie', 40))
tuple_df = pd.DataFrame(my_tuple, columns = ['Names', 'Age'], index = [1, 2, 3])
tuple_df
````

````python

my_dict = {
    "Names": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 40]
}

dict_df = pd.DataFrame(my_dict, index=[1, 2, 3]) # Key as a columns
dict_df
````

````python
my_dict = [
     {"Names": "Alice", "Age": 25},
     {"Names": "Bob", "Age": 30},
     {"Names": "Charlie", "Age": 40, "City" : "Dhaka"}
]

dict_df = pd.DataFrame(my_dict)
dict_df
````
