# 🔹 NumPy (Numerical Python)

**NumPy** is a powerful Python library used for numerical computing and working with arrays.

---

## 🔹 Key Features
- Provides fast and efficient **array operations**  
- Supports **multi-dimensional arrays**  
- Includes mathematical and statistical functions  
- Optimized for performance (faster than lists)  

---

## 🔹 Core Concept
- **ndarray (N-dimensional array)** → Main data structure in NumPy  
- Stores elements of the same data type  
- Allows vectorized operations  

---

## 🔹 Common Uses
- Data analysis  
- Scientific computing  
- Machine learning  
- Matrix operations  

---

## 🔹 Advantages
- Faster computation  
- Less memory usage  
- Easy mathematical operations  
- Supports broadcasting  

---

## 🔹 NumPy vs Python List

| Feature        | NumPy Array        | Python List       |
|----------------|------------------|------------------|
| Speed          | Faster            | Slower           |
| Memory         | Efficient         | More usage       |
| Data Type      | Same type         | Mixed types      |
| Operations     | Vectorized        | Manual loops     |

## Nd Array

```python
import numpy as np
# 1D Array:
arr1 = np.array([1,2,3,4,5])
print(type(arr1))
print(arr1.ndim)

# 2D Array

arr2 = np.array([[1,2,3], # Row 1
                 [4,5,6] # Row 2
                ])
print(arr2.ndim)

# 3D Array - ekta Array er vitor onkgula 2D array

arr3 = np.array([
        [ # 1st Floor
        [1,2,3],
        [4,5,6] 
        ],
        [ #2nd Floor
        [1,2,3], 
        [4,5,6]
        ]
                ]
               )
print(arr3.ndim)
```

## Array Attributes

```python
## Array-dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

# Shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# Data-Type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

# Size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

## ndarray creation from scratch

```python
# np.zeroes
arr = np.zeros((2,3), dtype = np.int8) # With row and column
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

arr = np.zeros_like(arr3) # In the first we take a 3d array
print(arr)
print(arr.shape)
```

```python
# np.ones
arr = np.ones((2,3), dtype = np.int8) # With row and column
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

arr = np.ones_like(arr3) # In the first we take a 3d array
print(arr)
print(arr.shape)
```

```python
# Empty
arr = np.empty((2,3), dtype = np.int8) # With row and column any data can occurs
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

arr = np.empty_like(arr3) # In the first we take a 3d array
print(arr)
print(arr.shape)
```

```python
# np.full
arr = np.full((2,3),np.inf)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

arr = np.full_like(arr3, np.inf, dtype = np.float64) # In the first we take a 3d array which data type was int64 bit
print(arr)
print(arr.shape)
```

## 🔹 linspace vs logspace

| Feature  | `linspace` 🟦    | `logspace` 🟧             |
|----------|------------------|---------------------------|
| Spacing  | Linear           | Logarithmic               |
| Pattern  | Equal difference | Equal ratio (multiplying) |
| Use case | Normal ranges    | Exponential/log scales    |
| Example  | 1, 2, 3, 4       | 10, 100, 1000             |

```python
# np.linespace(stary, end(included), total_point)
arr = np.linspace(1, 10, 10) # -> divide the line equally according to total_point
print(arr)

arr = np.linspace(1, 10, 11) # -> divide the line equally according to total_point
print(arr)


```

```python
# np.logspace(start, end(included), total_point)
arr = np.logspace(1, 4, 5) # -> 10 to the power point
print(arr)

arr = np.logspace(1, 4, 5, base = 2) # -> 2 to the power point
print(arr)
```

## Creating Diagonal and Identity Matrix

## 🔹 Diagonal vs Identity Matrix (Difference)

| Feature         | Diagonal Matrix 🟦        | Identity Matrix 🟧        |
|-----------------|---------------------------|---------------------------|
| Diagonal values | Any number                | Always 1                  |
| Type            | General matrix            | Special case of diagonal  |
| Example         | 5, 3, 8 (on diagonal)     | 1, 1, 1 (on diagonal)     |

```python
diagonal_matrix = np.diag([1,2,3])
print(diagonal_matrix)
```

```python
identity_matrix = np.eye(3)
print(identity_matrix)

# np.eye(row, column, k)
identity_matrix = np.eye(3, 4, -1) # -> for -1 it will shift to lower anf for 1 it will shift to upper
print(identity_matrix)
```
