# 🔹 EDA (Exploratory Data Analysis) Methods

EDA methods are techniques used to explore, summarize, and understand data before analysis or machine learning.

---

## 🔹 1. Univariate Analysis
Analysis of a single variable.

### ✔ Purpose
- Understand distribution
- Find central tendency and spread

### ✔ Common Methods
- Histogram
- Box Plot
- Mean, Median, Mode

---

## 🔹 2. Bivariate Analysis
Analysis of the relationship between two variables.

### ✔ Purpose
- Identify correlation or patterns

### ✔ Common Methods
- Scatter Plot
- Line Plot
- Correlation Analysis

---

## 🔹 3. Multivariate Analysis
Analysis of more than two variables together.

### ✔ Purpose
- Understand complex relationships
- Compare multiple features

### ✔ Common Methods
- Heatmap
- Pair Plot
- Grouped Analysis

---

## 🔹 4. Statistical Analysis
Uses numerical summaries to understand data.

### ✔ Common Methods
- Mean
- Standard Deviation
- Variance
- Correlation

---

## 🔹 5. Data Cleaning
Improves data quality before analysis.

### ✔ Tasks
- Handle missing values
- Remove duplicates
- Fix incorrect data types

---

## 🔹 Summary

EDA methods help analyze data from different perspectives to discover patterns, relationships, and data quality issues.

## Univariate Analysis - Analysis With a Specific Column

# 1. Categorical Values

```python
sns.countplot(data=df,x ="Survived")
```

```python
(df['Survived'].value_counts() / len(df)) * 100
```

```python
survived_count = df['Survived'].value_counts()
survived_label = df['Survived'].unique()
print(survived_label)
plt.pie(survived_count,labels =survived_label, autopct="%1.1f")

plt.title("Survival Rate")
plt.show()
```
